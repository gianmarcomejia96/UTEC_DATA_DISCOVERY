```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/"
print(f"\n1. URL objetivo: {url}")
response = requests.get(url)

print(f"\n2. Código de respuesta: {response.status_code}")

print(f"   Tamaño del contenido: {len(response.content)} bytes")

print(f"\n4. Texto del HTML:")
print(response.text)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
# print(soup.prettify())

title = soup.find("title")
print(f"\nTítulo de la página: {title.text}")

h1 = soup.find("h1")
if h1:
    print(f"\nPrimer encabezado H1: {h1.text}")
```

```python
s1 = soup.find("small")
if s1:
    print(s1.text)
```

```python
f1 = soup.find("form")
if f1:
    print(f1.text)
```

```python
products = soup.find_all("article")

data = []
for product in products :
    descrip = product.find("a", href =True,title=True).get_text(strip=True)
    p = product.find("p",class_="price_color").get_text(strip=True)
    data.append({"producto": descrip, "precio": p})

print(data)
```

```python
from urllib.parse import urljoin
url = "https://books.toscrape.com/"

data = []
while True:
    soup = BeautifulSoup(requests.get(url).text, "html.parser")
    products = soup.find_all("article")

    for product in products:
        descrip = product.find("a", href=True, title=True).get_text(strip=True)
        p = product.find("p", class_="price_color").get_text(strip=True)
        data.append({"producto": descrip, "precio": p})

    next_tag = soup.select_one("li.next a")
    if not next_tag:
        break
    url = urljoin(url, next_tag["href"])
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)

df
```

```python
df["precio"] = (
    df["precio"]
    .str.replace(r"[^0-9.]", "", regex=True)
    .astype(float)
)
```

```python
wait = WebDriverWait(driver, 10)
data = []

while True:
    wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "article.product_pod")))

    products = driver.find_elements(By.CSS_SELECTOR, "article.product_pod")

    for product in products:
        title = product.find_element(By.CSS_SELECTOR, "h3 a").get_attribute("title")
        price = product.find_element(By.CSS_SELECTOR, ".price_color").text
        data.append({
            "producto": title,
            "precio": price
        })

    try:
        next_button = wait.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "li.next > a"))
        )
        next_button.click()
    except:
        break

driver.quit()

df = pd.DataFrame(data)
print(df.head())
print(df.shape)

#df.to_csv("selenium_quotes_all_pages.csv", index=False, encoding="utf-8")
#print("Datos guardados en selenium_quotes_all_pages.csv")
```
